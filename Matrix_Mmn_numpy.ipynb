```python
import numpy as np
from fractions import Fraction

Q  = 9

#c0,c1,c2,c3,c4,c5,c6,c7,c8
cx = [0, 1, 0, -1, 0, 1, -1, -1, 1]
cy = [0, 0, 1, 0, -1, 1, 1, -1, -1]


#c0,c1,c2,c5,c3,c4,c6,c8,c7
#cx = [0, 1, -1, 0, 1, -1, 0, 1, -1]
#cy = [0, 0, 0, 1, 1, 1, -1, -1, -1]

pow_x = [0, 1, 0, 1, 2, 0, 2, 1, 2] 
pow_y = [0, 0, 1, 1, 0, 2, 1, 2, 2] 

th,ux, uy = 1, 1,1

T  = 1
TL = 4./3
th =1
M = np.zeros((Q, Q))

for i in range(Q):
    for j in range(Q):
        M[i, j] = (np.sqrt(th) * cx[j] + ux) ** pow_x[i] * (np.sqrt(th) * cy[j] + uy) ** pow_y[i]
```

```python
M
```

```python
M_inv = np.linalg.inv(M)


for row in M_inv:
    fraction_row = [str(Fraction(x).limit_denominator()) for x in row]
    print(' '.join(fraction_row))
```
